```python
import numpy as np

data = np.loadtxt('m_eta_0.8.txt') # return (N, 2) array
ts08 = data[:, 0]                       # return only 1st col
ms08 = data[:, 1]                       # return only 2nd col
data = np.loadtxt('r_eta_0.8.txt')
rs08 = data[:, 1]                       # data in Rsun

data = np.loadtxt('m_eta_0.6.txt')
ts06 = data[:, 0]
ms06 = data[:, 1]
data = np.loadtxt('r_eta_0.6.txt')
rs06 = data[:, 1]

data = np.loadtxt('m_eta_0.4.txt')
ts04 = data[:, 0]
ms04 = data[:, 1]
data = np.loadtxt('r_eta_0.4.txt')
rs04 = data[:, 1]

data = np.loadtxt('m_eta_0.2.txt')
ts02 = data[:, 0]
ms02 = data[:, 1]
data = np.loadtxt('r_eta_0.2.txt')
rs02 = data[:, 1]
```

```python
# %matplotlib widget

import matplotlib.pyplot as plt
import matplotlib.ticker as mticker

f = mticker.ScalarFormatter(useOffset=False, useMathText=True)
g = lambda x,pos : "${}$".format(f._formatSciNotation('%1.10e' % x))
props = dict(boxstyle='round', facecolor='wheat', alpha=0.5)

fig, (ax1, ax2) = plt.subplots(nrows=2, ncols=1, sharex=True, figsize=(8, 12),
                               dpi=300)
# fig, (ax1, ax2) = plt.subplots(nrows=2, ncols=1, sharex=True)
fig.subplots_adjust(hspace=0)

ax1.set_ylabel("$M(t)$ / $M_{\odot}$", fontsize='large')
ax1.yaxis.set_minor_locator(mticker.AutoMinorLocator())
ax1.plot(ts02, ms02, color='tab:red', label='$\eta = 0.2$')
ax1.plot(ts04, ms04, color='tab:orange', label='$\eta = 0.4$')
ax1.plot(ts06, ms06, color='tab:green', label='$\eta = 0.6$')
ax1.plot(ts08, ms08, color='tab:blue', label='$\eta = 0.8$')
ax1.text(0.025, 0.075, '(a)', transform=ax1.transAxes, fontsize='xx-large',
        verticalalignment='top')
ax1.grid()

ax2.set_ylabel('$R(t)$ / $R_{\odot}$', fontsize='large')
ax2.yaxis.set_minor_locator(mticker.AutoMinorLocator())
# ax2.ticklabel_format(axis='x', style='sci', scilimits=(0,0))
ax2.tick_params(axis='x', which='major')
ax2.xaxis.set_minor_locator(mticker.AutoMinorLocator())
ax2.set_xlabel('Age / yr', fontsize='large')
ax2.plot(ts02, rs02, color='tab:red', label='$\eta = 0.2$')
ax2.plot(ts04, rs04, color='tab:orange', label='$\eta = 0.4$')
ax2.plot(ts06, rs06, color='tab:green', label='$\eta = 0.6$')
ax2.plot(ts08, rs08, color='tab:blue', label='$\eta = 0.8$')
ax2.set_xlim(1.230e10, 1.255e10)
ax2.legend(fontsize='large', loc='upper right')
ax2.text(0.025, 0.075, '(b)', transform=ax2.transAxes, fontsize='xx-large',
        verticalalignment='top')
ax2.grid()

plt.gca().xaxis.set_major_formatter(mticker.FuncFormatter(g))
plt.savefig('eta_comparison.png')
```
